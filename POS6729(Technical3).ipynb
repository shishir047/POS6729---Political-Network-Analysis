```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from networkx.algorithms import community
from collections import Counter

from google.colab import files
uploaded = files.upload()

cooperation_1 = nx.read_edgelist('cooperation.txt', nodetype=int, data=(('weight',float),))

conflict_df = pd.read_csv('conflict.csv', index_col=0)
conflict_1 = nx.from_pandas_adjacency(conflict_df)
```

```python
def calculate_network_metrics(G):
    metrics = {
        'Number of Nodes': G.number_of_nodes(),
        'Number of Edges': G.number_of_edges(),
        'Average Degree': np.mean([d for n, d in G.degree()]),
        'Density': nx.density(G),
        'Average Clustering Coefficient': nx.average_clustering(G),
        'Is Connected': nx.is_connected(G),
    }
    if metrics['Is Connected']:
        metrics['Average Shortest Path Length'] = nx.average_shortest_path_length(G)
        metrics['Diameter'] = nx.diameter(G)
    else:
        metrics['Average Shortest Path Length'] = 'N/A'
        metrics['Diameter'] = 'N/A'

    degrees = [d for n, d in G.degree()]
    metrics['Gini Coefficient'] = gini(degrees)

    communities = community.greedy_modularity_communities(G)
    metrics['Number of Communities'] = len(communities)
    return metrics

def gini(x):
    mad = np.abs(np.subtract.outer(x, x)).mean()
    rmad = mad / np.mean(x)
    g = 0.5 * rmad
    return g
```

```python
cooperation_metrics = calculate_network_metrics(cooperation_1)
conflict_metrics = calculate_network_metrics(conflict_1)

print("Cooperation Network Metrics:")
for k, v in cooperation_metrics.items():
    print(f"{k}: {v}")

print("\nConflict Network Metrics:")
for k, v in conflict_metrics.items():
    print(f"{k}: {v}")
```
